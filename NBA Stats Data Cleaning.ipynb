## Cleaning NBA Stats From Webscraping

```python
import pandas as pd
import numpy as np
```

```python
mvp = pd.read_csv('mvps.csv')
```

```python
mvp.head()
```

## Data Cleaning

```python
mvp = mvp[['Player', 'Year', 'Pts Won', 'Pts Max', 'Share']]
```

```python
mvp.head()
```

```python
player = pd.read_csv('players.csv')
```

```python
player.head()
```

```python
del player['Unnamed: 0']
del player['Rk']
```

```python
player.head()
```

Some players in the player column contain an * after their name. This needs to be removed since the tables have to be merged. 

```python
player['Player'] = player['Player'].str.replace('*', '', regex = False)
```

Some players have more than one entry  in this dataframe. They may have been traded mid season or have joined  a different  team  after their contract expired. These players need to have their entries  merged into one column.

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else: 
        row = df[df['Tm'] == 'TOT']
        df['Tm'] = df.iloc[-1,:]['Tm']
        
player = player.groupby(['Player', 'Year']).apply(single_row)
```

```python
player.index = player.index.droplevel()
player.index = player.index.droplevel()
```

## Combining Player and MVP data

```python
combined = player.merge(mvp, how = 'outer', on =['Player', 'Year'])
```

```python
combined[['Pts Won',  'Pts Max', "Share"]] = combined[['Pts Won',  'Pts Max', "Share"]].fillna(0)
```

```python
combined.head()
```

## Cleaning the Team Data


```python
team = pd.read_csv('teams.csv')
```

```python
team = team[~team['W'].str.contains('Division')]
```

```python
team.head()
```

```python
team['Team'] = team['Team'].str.replace('*', '', regex = False)
```

```python
nicknames = pd.read_csv('nicknames.csv')
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
stats = combined.merge(team, how = 'outer', on = ['Team','Year'])
```

```python
stats.head()
```

```python
del stats['Unnamed: 0']
```

```python
stats = stats.apply(pd.to_numeric, errors = 'ignore')
```

```python
stats.info()
```

```python
stats.to_csv('player_mvp_stats.csv')
```
