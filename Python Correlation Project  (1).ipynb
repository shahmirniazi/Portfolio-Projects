```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8)

#read in the data 

df = pd.read_csv('/Users/macuser/Downloads/projects/Python Project/movies.csv')
```

```python
#looking at the data using the head function
df.head()
```

```python
# finding any missing data or null values 
df = df.dropna()
for col in df.columns:
    prct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, prct_missing))
```

```python
#data types for columns 
df.dtypes

```

```python
#converting budget and gross to int values 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False  )
```

```python
pd.set_option('display.max_rows', None)
```

```python
#checking to see if there are any duplicates
#no duplicates present

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
df.head()
```

```python
# budget will have a high correlation with revenue (Correct)
# company will have a high correlation with revenue (Incorrect)

```

```python
#building a scatterplot of budget vs gross revenue 

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel(' Budget for film')
plt.ylabel('Gross earnings ')
plt.show()
```

```python
df.head()
```

```python
# make a regression plot 

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color":"green"}, line_kws= {"color":"blue"})
```

```python
#finding the exact correlation

df.corr()
```

```python
#High correlation between budget and gross revenue

```

```python
corrmatrix = df.corr()
sns.heatmap(corrmatrix, annot = True )
plt.title('Correlation Matrix')
plt.xlabel('Film Parameters')
plt.ylabel('Film Parameters')
plt.show
```

```python
#seeing the correlation between company and gross revenue 
#Company is a string so we need a numeric representation of it 
```

```python
df_numerized = df 

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
    
```

```python
corrmatrix = df_numerized.corr()
sns.heatmap(corrmatrix, annot = True )
plt.title('Correlation Matrix')
plt.xlabel('Film Parameters')
plt.ylabel('Film Parameters')
plt.show
```

```python
#No significant correlation between company and gross revenue 
#The only significant corrlation is between budget and gross revenue, and votes and gross revenue

```

```python
#to get a numerical view of the correlations 
corr_matrix = df_numerized.corr()
pairs = corr_matrix.unstack()
pairs.head()
```

```python
sorted_pairs = pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

```python
#Summary: Budget and Votes have the highest correlation to Gross Revenue of a movie
```
